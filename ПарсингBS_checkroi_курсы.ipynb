```python
#получить исходный код страницы

import requests
from bs4 import BeautifulSoup

# Метод получения HTML указанной страницы (исходного кода)
def get_html():
# Создание объекта сессии
    session = requests.session()             
# Добавление заголовков к сессии присвоили заголовки к сессии.
    session.headers = {
      'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.143 YaBrowser/22.5.0.1816 Yowser/2.5 Safari/537.36'
  }
    # объявили переменную, страницу, с которой сделать запрос
    url = 'https://checkroi.ru/napravlenie/programmirovanie/'
    try:
        # получаем ихсодный код страницы, через session.get
        res = session.get(url)
        # делаем проверку   
        res.raise_for_status()
    except Exception as ex:
        print(ex)
        return
        
    return res.text
html = get_html()   
soup = BeautifulSoup(html, 'lxml')
```

```python
from bs4 import BeautifulSoup


'''данная функция собирает HTML'''
def get_html(page):
    session = requests.session()             # Создание объекта сессии
    session.headers = {
      'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.143 YaBrowser/22.5.0.1816 Yowser/2.5 Safari/537.36'
      'accept-language: ru,en;q=0.9'  
  }
   
    url = 'https://checkroi.ru/napravlenie/programmirovanie/page/{}/'
    try:
        # получаем ихсодный код страницы, через session.get
        res = session.get(url.format(page))
        # делаем проверку   
        res.raise_for_status()
    except Exception as ex:
        print(ex)
        return
        
    return res.text

'''данная функция варит суп'''
def get_soup(page):
    html = get_html(page)
    soup = BeautifulSoup(html, 'lxml')
    return soup

'''собирает информацию о курсе'''
def get_course_info(first_course):
    title_course = first_course.select_one('h3 a').text
    if not title_course:
        print('error find title')
        return
    duration_course = first_course.select_one('div.woo_code_zone_loop').text
    if not duration_course:
        print('error find duration')
        return
    rate_course = first_course.select_one('div.overall-score').text 
    if not rate_course:
        print('error find rating')
        return
        
    price_course = first_course.select_one('span.price').text 
    if not price_course:
        print('error find price')
        return
        
    return {
        'title': title_course,
        'duration': duration_course,
        'rating': rate_course,
        'price': price_course
    }
    
'''собирает информацию о курсах на странице'''
def get_courses_info(soup):
    course_rows = soup.select_one('div.columns-3') 
    all_courses = course_rows.select('div.product')   
    result = []                         # записываем в словарть              
    for div in all_courses:
      result.append(get_course_info(div))
    return result
```

```python
course_rows = soup.select_one('div.columns-3') 
all_courses = course_rows.select('div.product')   
result = []
for div in all_courses:
    result.append(get_course_info(div))
    #print(result)
    print(get_course_info(div))
```

```python
#1. цикл по страница 1--6
#2. получить исходный код
#3. сварить суп из исходного кода
#4. получить инфо о курсах
#5. for page in range (1, 51) будем выполнять наш код
```

```python
import time
import csv 
result = []
for page in range (1, 19):
    time.sleep(3)   # спит сколько секунд
    soup = get_soup(page)
    courses_info = get_courses_info(soup)
    result = result + courses_info
    print(result)
    
```

```python
with open('courses.csv', 'w') as f:    
    writer = csv.DictWriter(f, result[0].keys())
    writer.writeheader()
    for r in result:
        writer.writerow(r)
    #print(result)
```
